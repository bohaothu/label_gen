```python
import numpy as np
import pandas as pd
import ujson
```

```python
row_num, feature_num, label_num = 100, 5, 30
fake_features=np.random.rand(row_num, feature_num)
df=pd.DataFrame(data=fake_features,columns=[str("feature")+str(i+1) for i in range(feature_num)])
```

```python
def gen_matrix_col(feature_n):
    z = np.random.rand(1,feature_n)
    z = z/np.sum(z)
    return np.transpose(z)
def gen_matrix(feature_n, label_n):
    return np.concatenate([ gen_matrix_col(feature_n) for x in range(label_n) ],axis=1)
gen_mat = gen_matrix(feature_num,label_num)
```

```python
def map_value(n):
    return 1 if n > 0.5 else 0
def gen_label(arr):
    return np.array([map_value(y) for y in np.dot(arr,gen_mat)])
```

```python
df["labels"]=df.apply(lambda x: gen_label(x), axis=1)
df["labels"]=df["labels"].apply(lambda x: str(ujson.dumps(x.tolist())))
```
